```python
## Numpy
import numpy as np
import time
```

```python
# Creating array from list

array_1d = np.array([1, 2, 3, 4, 5, 6])  # 1D array
print("1D Array: ", array_1d)

array_2d = np.array([[1, 2, 3], [4, 5, 6]])  # 2D array
print("2D Array: ", array_2d)
```

```python
# List vs Numpy

py_list = [1, 2, 3, 4, 5]
print("Multiplication of list:", py_list * 2)

np_array = np.array([1, 2, 3])  # element wise multiplication
print("Multiplication of Numpy Array: ", np_array * 2)
import time

start = time.time()  # start time

py_list = [i * 2 for i in range(10000)]

end = time.time()  # end time

print("List operation time:", end - start)
```

```python
# Creating array from scratch

zeros = np.zeros((3, 4))
print(zeros)

ones = np.ones((2, 3))
print(ones)

full = np.full((2, 2), 7)
print(full)

random = np.random.random((2, 3))
print(random)

sequence = np.arange(0, 10, 2)
print(sequence)
```

```python
# Vector, Matrix, Tensor

vector = np.array([1, 2, 3, 4])
print(vector)

matrix = np.array([[1, 2],
                   [3, 4]])
print(matrix)

tensor = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(tensor)
```

```python
# Array properties

arr = np.array([
    [1, 2],
    [3, 4]
])
print(arr.shape)  # rows and column
print(arr.ndim)  # gives dimension
print(arr.size)  # gives total number of element
print(arr.dtype)  # gives data type
```
